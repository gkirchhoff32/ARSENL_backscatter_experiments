# Generate Simulated Data

```python
# simulation resolution settings
t_sim_min = 0
t_sim_max = 40e-9
dt_sim = 25e-12

tD = 25e-9 # deadtime
Nshot = int(1e4) # number of laser shots
wrap_deadtime = True  # wrap deadtime between shots

window_bnd = [28e-9, 34e-9]

laser_pulse_width = 700e-12 # laser pulse width in seconds --> FWHM = 2.355*sigma
target_time = 31.2e-9
target_amplitude = 1e8 # target peak count rate
background = 1e1  # background count rate
```

```python
# generate the simulated scene time axis
t_sim = np.arange(t_sim_min,t_sim_max,dt_sim)           # simulation time
t_sim_bins = np.concatenate((t_sim,t_sim[-1:]+dt_sim))  # simulation time histogram bins
```

```python
# generate the photon arrival rate of the profile
# Gaussian target with constant background
photon_rate_arr = target_amplitude*np.exp(-(t_sim - target_time)**2/(2*laser_pulse_width**2))+background
```

```python
# generate photon counts

# lists of photon arrivals per laser shot
start = time.time()
t_det_lst = []  # detected photons (includes deadtime)
t_phot_lst = [] # actual photons (no dead time)

t_det_last = -100.0  # last photon detection event
for n in range(Nshot):
    # simulate a laser shot
    ptime, ctime = sim.photon_count_generator(t_sim_bins, 
                                              photon_rate_arr, 
                                              tau_d_flt=tD,
                                              last_photon_flt=t_det_last)
    if wrap_deadtime:
        if len(ctime) > 0:
            t_det_last = ctime[-1]
        t_det_last -= t_sim_bins[-1]
        
    t_det_lst += [ctime]  # detection time tags (including deadtime)
    t_phot_lst += [ptime] # photon time tags

print('time elapsed: {}'.format(time.time()-start))
```

```python
phot_arr = np.array(sorted(np.concatenate(t_phot_lst)))
plt.figure()
plt.stem(phot_arr,np.ones(phot_arr.size))
plt.title('Photons')

cnt_arr = np.array(sorted(np.concatenate(t_det_lst)))
plt.figure()
plt.stem(cnt_arr,np.ones(cnt_arr.size))
plt.title('Detected Photons')
```

```python
flight_time = xr.DataArray(np.concatenate(t_det_lst))
# flight_time = flight_time[np.where((flight_time>=window_bnd[0]) & (flight_time<window_bnd[1]))]  # Exclude specified t.o.f. bins
n_shots = Nshot
t_det_lst = t_det_lst

true_flight_time = xr.DataArray(np.concatenate(t_phot_lst))
# true_flight_time = true_flight_time[np.where((true_flight_time>=window_bnd[0]) & (true_flight_time<window_bnd[1]))]
```

```python
bin_edges = t_sim_bins[::1]
active_ratio_hst = Nshot * np.ones(len(bin_edges)-1)
deadtime_n_bins = np.floor(tD / dt_sim).astype(int)  # Number of bins that deadtime occupies

for shot in range(len(t_det_lst)):
    detections = t_det_lst[shot]

    for det in detections:
        det_time = det.item()  # Time tag of detection that occurred during laser shot

        # Only include detections that fall within fitting window
        if det_time >= (t_sim_min-tD) and det_time <= t_sim_max:
            det_bin_idx = np.argmin(abs(det_time - bin_edges))  # Bin that detection falls into
            final_dead_bin = det_bin_idx + deadtime_n_bins  # Final bin index that deadtime occupies

            # Currently a crutch that assumes "dead" time >> potwindow. Will need to include "wrap around" to be more accurate
            # If final dead bin surpasses fit window, set it to the window upper bin
            if final_dead_bin > len(active_ratio_hst):
                final_dead_bin = len(active_ratio_hst)
            # If initial dead bin (detection bin) precedes fit window, set it to the window lower bin
            if det_time < t_sim_min:
                det_bin_idx = 0
            active_ratio_hst[det_bin_idx:final_dead_bin+1] -= 1  # Remove "dead" region in active ratio

active_ratio_hst /= n_shots  # Normalize for ratio
```

```python
fig = plt.figure(figsize=(18,10), dpi=400)
plt.rcParams.update({'font.size': 18})

# ax1 = fig.add_subplot(411)
# phot_arr = np.array(sorted(np.concatenate(t_phot_lst)))
# ax1.stem(phot_arr,np.ones(phot_arr.size))
# ax1.set_title('Photons')
# ax1.set_xlim([window_bnd[0], window_bnd[1]])

crop_num = 200

ax2 = fig.add_subplot(311)
cnt_arr = np.array(sorted(np.concatenate(t_det_lst)[:crop_num]))
ax2.stem(cnt_arr*1e9,0.5*np.ones(cnt_arr.size), basefmt=" ", linefmt='k-', markerfmt='ko')
# ax2.set_title('First {:.0f} Detected Photons'.format(crop_num))
ax2.set_xlim([window_bnd[0]*1e9, window_bnd[1]*1e9])
ax2.set_ylim([0, 0.75])
ax2.get_yaxis().set_visible(False)

# Scaled time-of-flight histogram
ax3 = fig.add_subplot(312)

# bin_array = np.arange(window_bnd[0], window_bnd[1]+dt_sim, dt_sim)
bin_array = bin_edges

n, bins = np.histogram(flight_time, bins=bin_array)
binwidth = np.diff(bins)[0]
N = n / binwidth / n_shots  # [Hz] Scaling counts to arrival rate
center = 0.5 * (bins[:-1] + bins[1:])
ax3.bar(center*1e9, n, align='center', width=binwidth*1e9, color='b', alpha=0.5)
ax3.set_ylabel(r'$\bf{Y}$')
ax3.set_xlim([window_bnd[0]*1e9, window_bnd[1]*1e9])

ax4 = fig.add_subplot(313)
ax4.bar(center*1e9, active_ratio_hst, align='center', width=binwidth*1e9, color='purple', alpha=0.5)
ax4.set_xlabel('Time of flight [ns]')
ax4.set_xlim([window_bnd[0]*1e9, window_bnd[1]*1e9])
ax4.set_ylim([0.99*min(active_ratio_hst), 1.01*max(active_ratio_hst)])
ax4.set_ylabel(r'$\bf{Z}$')

plt.tight_layout()
plt.show()
```

Double check the integral approximation using erf()

```python
t_torch = torch.tensor(t_sim)
integral_est = torch.tensor(t_sim*background) \
        + 0.5*target_amplitude*laser_pulse_width*np.sqrt(2*np.pi) \
        + 0.5*target_amplitude*laser_pulse_width*np.sqrt(2*np.pi)\
            *torch.special.erf((t_torch-target_time)/(laser_pulse_width*np.sqrt(2)))
```
